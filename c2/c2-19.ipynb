```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
import librosa

try:
    x, Fs = librosa.load('tenki.mp3', sr=None, mono=True)
    print(f"đi dạo với {len(x)} samples, {Fs} Hz")
except:
    Fs = 8000
    duration = 3
    t = np.linspace(0, duration, int(Fs * duration))
    
    x = (np.sin(2*np.pi*440*t) * np.exp(-t/0.5) +
         np.sin(2*np.pi*554*t) * np.exp(-(t-0.3)/0.5) * (t > 0.3) +
         np.sin(2*np.pi*659*t) * np.exp(-(t-0.6)/0.5) * (t > 0.6))
    
    x = x / np.max(np.abs(x))
    print(f"không đi dạo được tạo sound test: {len(x)} samples, {Fs} Hz")

plt.figure(figsize=(12, 3))
plt.plot(x[:Fs])
plt.title('đi dạo gốc')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"\nAudio gốc:")
Audio(x, rate=Fs)
```

```python
#τ = 50 ms, a = 0.7
tau_a = 0.050
a_a = 0.7
D_a = int(tau_a * Fs)

print(f"τ = {tau_a*1000} ms, a = {a_a}, D = {D_a} samples")
print(f"y[n] = x[n] + {a_a}·y[n - {D_a}]")

y_a = np.zeros(len(x))
for n in range(len(x)):
    if n < D_a:
        y_a[n] = x[n]
    else:
        y_a[n] = x[n] + a_a * y_a[n - D_a]

y_a = y_a / np.max(np.abs(y_a))

plt.figure(figsize=(12, 4))
plt.subplot(2, 1, 1)
plt.plot(x[:Fs], label='Gốc')
plt.title('Tín hiệu gốc')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(y_a[:Fs], label='τ=50ms', color='green')
plt.title('Vọng 1')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("các âm vọng gần, nhỏ, reverb nhẹ, nghe giống mấy hồi bị lỗi âm thanh trên sân khấu")
Audio(y_a, rate=Fs)
```
