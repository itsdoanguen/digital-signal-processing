```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

h = np.array([1, -2, 1], dtype=np.float32)
print("Impulse response h[n]:", h)
print("This represents: h[-1]=1, h[0]=-2, h[1]=1")
```

```python
#theo hàng
img = cv2.imread('wonderacute.jpg', cv2.IMREAD_GRAYSCALE)

if img is not None:
    print(f"Image loaded successfully with shape: {img.shape}")
    
    h_row = h.reshape(1, 3)
    img_row_edges = cv2.filter2D(img, -1, h_row)
    
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Ảnh gốc')
    plt.axis('off')
    
    plt.subplot(1, 3, 2)
    plt.imshow(img_row_edges, cmap='gray')
    plt.title('Xử lý theo hàng (Raw)')
    plt.axis('off')
    
    plt.subplot(1, 3, 3)
    plt.imshow(np.abs(img_row_edges), cmap='gray')
    plt.title('Xử lý theo hàng (Absolute)')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
else:
    print("Ảnh?")
```

```python
#theo cột
if img is not None:
    h_col = h.reshape(3, 1)
    img_col_edges = cv2.filter2D(img, -1, h_col)
    
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Ảnh gốc')
    plt.axis('off')
    
    plt.subplot(1, 3, 2)
    plt.imshow(img_col_edges, cmap='gray')
    plt.title('Xử lý theo cột (Raw)')
    plt.axis('off')
    
    plt.subplot(1, 3, 3)
    plt.imshow(np.abs(img_col_edges), cmap='gray')
    plt.title('Xử lý theo cột (Absolute)')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
else:
    print("Ảnh?")
```

```python
#ssanh
if img is not None:
    img_combined = np.sqrt(img_row_edges**2 + img_col_edges**2)
    
    plt.figure(figsize=(20, 10))
    
    plt.subplot(2, 3, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Ảnh gốc', fontsize=14)
    plt.axis('off')
    
    plt.subplot(2, 3, 2)
    plt.imshow(img_row_edges, cmap='gray')
    plt.title('Theo hàng (Raw) - Phát hiện cạnh dọc', fontsize=14)
    plt.axis('off')
    
    plt.subplot(2, 3, 3)
    plt.imshow(np.abs(img_row_edges), cmap='gray')
    plt.title('Theo hàng (Absolute) - Phát hiện cạnh dọc', fontsize=14)
    plt.axis('off')
    
    plt.subplot(2, 3, 5)
    plt.imshow(img_col_edges, cmap='gray')
    plt.title('Theo cột (Raw) - Phát hiện cạnh ngang', fontsize=14)
    plt.axis('off')
    
    plt.subplot(2, 3, 6)
    plt.imshow(np.abs(img_col_edges), cmap='gray')
    plt.title('Theo cột (Absolute) - Phát hiện cạnh ngang', fontsize=14)
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    
    # Show combined edge detection
    plt.figure(figsize=(10, 5))
    plt.imshow(img_combined, cmap='gray')
    plt.title('Kết hợp cả hai hướng (Magnitude)', fontsize=14)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
else:
    print("Ảnh?")
```
