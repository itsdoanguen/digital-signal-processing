```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('wonderacute.jpg', cv2.IMREAD_COLOR)

if img is not None:
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    plt.figure(figsize=(20, 10))
    
    #gốc
    plt.subplot(2, 3, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Ngựa gốc')
    plt.axis('off')
    
    #Sobel ngang - phát hiện cạnh dọc
    h_sobel_x = np.array([[1, 0, -1],
                          [2, 0, -2],
                          [1, 0, -1]], dtype=np.float32)
    
    img_sobel_x = cv2.filter2D(img_gray, -1, h_sobel_x)
    
    plt.subplot(2, 3, 2)
    plt.imshow(img_sobel_x, cmap='gray')
    plt.title('Ngựa ngang')
    plt.axis('off')
    
    plt.subplot(2, 3, 3)
    plt.imshow(np.abs(img_sobel_x), cmap='gray')
    plt.title('Ngựa ngang (Absolute)')
    plt.axis('off')
    
    #Sobel dọc - phát hiện cạnh ngang
    h_sobel_y = np.array([[1, 2, 1],
                          [0, 0, 0],
                          [-1, -2, -1]], dtype=np.float32)
    
    img_sobel_y = cv2.filter2D(img_gray, -1, h_sobel_y)
    
    plt.subplot(2, 3, 5)
    plt.imshow(img_sobel_y, cmap='gray')
    plt.title('Ngựa dọc')
    plt.axis('off')
    
    plt.subplot(2, 3, 6)
    plt.imshow(np.abs(img_sobel_y), cmap='gray')
    plt.title('Ngựa dọc (Absolute)')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    
else: 
    print("Ảnh đâu?")
```
