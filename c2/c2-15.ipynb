```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

image_path = 'wonderacute.jpg' 
img = cv2.imread(image_path, cv2.IMREAD_COLOR)

if img is not None:
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Ngựa gốc')
    plt.axis('off')

    # 3x3
    h3 = np.ones((3, 3), dtype=np.float32) / 9.0
    img_3x3 = cv2.filter2D(img, -1, h3)

    plt.subplot(1, 3, 2)
    plt.imshow(cv2.cvtColor(img_3x3, cv2.COLOR_BGR2RGB))
    plt.title('Ngựa 3x3')
    plt.axis('off')

    #10x10
    h10 = np.ones((10, 10), dtype=np.float32) / 100.0
    img_10x10 = cv2.filter2D(img, -1, h10)

    plt.subplot(1, 3, 3)
    plt.imshow(cv2.cvtColor(img_10x10, cv2.COLOR_BGR2RGB))
    plt.title('Ngựa 10x10')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
else:
    print(f"Ảnh đâu?")
```
