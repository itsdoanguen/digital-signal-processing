```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('wonderacute.jpg', cv2.IMREAD_COLOR)

if img is not None:
    plt.figure(figsize=(20, 5))
    
    plt.subplot(1, 4, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Ngựa gốc')
    plt.axis('off')
    
    #ngang
    h_row = np.ones((1, 10), dtype=np.float32) / 10.0
    img_row = cv2.filter2D(img, -1, h_row)
    
    plt.subplot(1, 4, 2)
    plt.imshow(cv2.cvtColor(img_row, cv2.COLOR_BGR2RGB))
    plt.title('Ngựa ngang')
    plt.axis('off')
    
    #dọc
    h_col = np.ones((10, 1), dtype=np.float32) / 10.0
    img_col = cv2.filter2D(img, -1, h_col)
    
    plt.subplot(1, 4, 3)
    plt.imshow(cv2.cvtColor(img_col, cv2.COLOR_BGR2RGB))
    plt.title('Ngựa dọc')
    plt.axis('off')
    
    #both
    img_row_then_col = cv2.filter2D(img_row, -1, h_col)
    plt.subplot(1, 4, 4)
    plt.imshow(cv2.cvtColor(img_row_then_col, cv2.COLOR_BGR2RGB))
    plt.title('Cả 2')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
else: 
    print("Ảnh đâu?")
```
